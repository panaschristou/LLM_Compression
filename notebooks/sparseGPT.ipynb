```python
def get_perplexity(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Reverse the lines to start reading from the end
    lines = lines[::-1]

    # Variables to store evaluation data
    last_evaluation = {}
    dataset_count = 0
    dataset_names = ['c4','ptb','wikitext2']
    length = [45576, 3761, 4358]
    for line in lines:
        # Assuming each dataset's evaluation starts with the word 'Dataset'
        if line.startswith('Perplexity'):
            dataset_info = line.split()
            #dataset_name = dataset_info[1]
            perplexity =  str(float(dataset_info[-1])) #/length[dataset_count]   # Assuming the perplexity is the last element
            last_evaluation[dataset_names[dataset_count]] = perplexity
            dataset_count += 1

            # Break the loop after reading three datasets
            if dataset_count == 3:
                break
    return last_evaluation
```

```python
def get_time(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    # Reverse the lines to start reading from the end
    lines = lines[::-1]

    # Variables to store evaluation data
    last_evaluation = {}
    dataset_count = 0
    dataset_names = ['c4','ptb','wikitext2']
    length = [45576, 3761, 4358]
    for line in lines:
        # Assuming each dataset's evaluation starts with the word 'Dataset'
        if line.startswith('Duration'):
            dataset_info = line.split()
            #dataset_name = dataset_info[1]
            perplexity =  str(float(dataset_info[-1])) #/length[dataset_count]   # Assuming the perplexity is the last element
            last_evaluation[dataset_names[dataset_count]] = perplexity
            dataset_count += 1

            # Break the loop after reading three datasets
            if dataset_count == 3:
                break
    return last_evaluation
```

```python
last_evaluations = {}
for ratio in np.sort(ratios):
    file_path = '/scratch/xc1490/projects/hpml/project/sparsegpt/output/txts/llama_{}.txt'.format(ratio)
    last_evaluations[ratio] = get_perplexity(file_path)
```

```python
df = pd.DataFrame.from_dict(last_evaluations, orient='index').astype(float)
df.index.name = 'ratio'
df.reset_index(inplace=True)
df['double_prune'] = np.array(['_' in i for i in df.ratio  ])
df['final_ratio'] = np.array([ '{:.2f}'.format(1-(1-float(i.split('_')[0]))**2) if '_' in i else i  for i in df.ratio ]).astype('float')
df['prune_ratio'] = np.repeat(np.array([  i.split('_')[0]    for i in df.ratio if '_' in i]).astype('float'),2)
```

```python
final_ratio = 0.8
with plt.style.context('ieee'):
    fig,ax = plt.subplots(1,2,figsize=(12, 4))
    # Plot for each dataset
    for ind, column in enumerate(df.columns[np.array([1,3])]):
        tmp_df = df[df['double_prune'] == True]
        ax[ind].plot(tmp_df[tmp_df['final_ratio'] <=final_ratio]['final_ratio'], \
                tmp_df[tmp_df['final_ratio'] <=final_ratio][column], marker='o', linestyle='-',label=column+' double prune')
        tmp_df = df[df['double_prune'] == False]
        ax[ind].plot(tmp_df[tmp_df['final_ratio'] <=final_ratio]['final_ratio'], \
                tmp_df[tmp_df['final_ratio'] <=final_ratio][column], marker='*', linestyle=':',label=column+' one pass prune')
    for i in range(2):
        ax[i].set_title('Sparse Pruning Evaluation Comparison')
        ax[i].set_xlabel('Equivalent Pruning Ratio')
        ax[i].set_ylabel('Perplexity')
        ax[i].legend()
        ax[i].grid(True)
    plt.show()
    fig.savefig('figs/sparseprune_{}_twice.png'.format(final_ratio))
```
